```python
#Questions
# 1) Difference in speed for the 50km,50mi male to female
```

```python
df3.groupby(['race_distance','athlete_gender'])['athlete_average_speed'].mean()
```

```python
# 2) what age groups are the best in the 50mi race (20+ races minimum )
```

```python
df3.query('race_distance == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending = False).query('count >=20')
```

```python
# 2) what age groups are the best in the 50mi race (10+ races minimum ) show 15
```

```python
df3.query('race_distance == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending = True).query('count >=10').head(15)
```

```python
# seasons for the data -> slower in summer than winter?
# spring 3-5
# summer 6-8
# fall 9-11
# winter 12-2
# split between two decimals
```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x : 'winter' if x>11 else 'fall' if x>8 else 'summer' if x>5 else 'spring' if x>2 else 'winter')
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending = False)
                                                                                
```

```python
df3.groupby(['race_season','athlete_gender'])['athlete_average_speed'].mean()
```
